```python
import pandas as pd
```

```python
all_participants_2019 = pd.read_csv('./challenge-raw-data/NAIC2019.csv',nrows=256)

columns_2019 = all_participants_2019.columns
num_of_participants_2019 = all_participants_2019.count()
print(num_of_participants_2019)
print(columns_2019)

# remove dollar sign and change the type to numeric
all_participants_2019['Final Award'] = all_participants_2019['Final Award'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# rename the column
all_participants_2019.rename(columns={'Final Award': 'Received MVP Award?'}, inplace=True)

# update the values 
all_participants_2019['Received MVP Award?'] = all_participants_2019['Received MVP Award?'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
winners_2019 = all_participants_2019[all_participants_2019['Received MVP Award?'] == 'Yes']
print(winners_2019)
```

```python
all_participants_2023 = pd.read_csv('./challenge-raw-data/NAIC2023.csv',nrows=223)

columns_2023 = all_participants_2023.columns
num_of_participants_2023 = all_participants_2023.count()
print(num_of_participants_2023)
print(columns_2023)

def clean_award_column(value):
    replaced_value = 0
    if isinstance(value, str):
        replaced_value = value.replace('$', '').replace(',', '').replace('.', '').replace('-','').replace(' ','')
        if (replaced_value == ''):
            return 0
        else:
            replaced_value = float(replaced_value)
    return replaced_value

# remove dollar sign and change the type to numeric
all_participants_2023['Received MVP Award?'] = all_participants_2023['Received MVP Award?'].apply(clean_award_column)

# update the values 
all_participants_2023['Received MVP Award?'] = all_participants_2023['Received MVP Award?'].apply(lambda x: 'Yes' if x > 0 else 'No')
```

```python
winners_2023 = all_participants_2023[all_participants_2023['Received MVP Award?'] == 'Yes']
print(winners_2023)
```

```python
# combining winners
winners_list = [winners_2019, winners_2020, winners_2021, winners_2022, winners_2023]
years = ['2019', '2020', '2021', '2022', '2023']

winners_total = pd.concat([winner.assign(Year=year) for winner, year in zip(winners_list, years)], ignore_index=True)
print(winners_total.columns)

unnecessary_columns = ['PI Last Name', 'REceiving $500 Check?', 'Phone #', 'Funded Prototype Development?', 'Funded MVP Development?', 'Pitched?', 'Submitted Prototype?', 'Submitted MVP?', 'Phonetic or Nickname', 'PSU Email', 'PSU ID', 'LinkedIn URL', 'Active Member?', 'Point of Initial Engagement', '2021 NAIC Participant?', '2020 NAIC Participant?', 'Graduation (if student)', 'Graduation Semester', 'Graduation Year', 'Receiving $500 Check?', 'Received MVP Award?',
                       'Social Media Handle', 'Role', 'Non PSU Email', 'Notes','Participating in Networking Event?','Funded Idea?', 'SME']
winners_total = winners_total.drop(columns = unnecessary_columns, errors='ignore')
print(winners_total.count())
winners_total.to_csv('./challenge-modified-data/winners_2019_2023.csv', index=False)
```
