```python
import pandas as pd

judge_1 = pd.read_csv("judge_1.csv", skiprows=3)
judge_2 = pd.read_csv("judge_2.csv", skiprows=3)
judge_3 = pd.read_csv("judge_3.csv", skiprows=3)
judge_4 = pd.read_csv("judge_4.csv", skiprows=3)
judge_5 = pd.read_csv("judge_5.csv", skiprows=3)
judge_6 = pd.read_csv("judge_6.csv", skiprows=3)
judge_7 = pd.read_csv("judge_7.csv", skiprows=3)


judge_1_clean = judge_1.drop(index = range(0,6)).drop(columns='Unnamed: 9', axis=1)
judge_2_clean = judge_2.drop(index = range(0,22)).drop(columns='Unnamed: 9', axis=1)
judge_3_clean = judge_3.drop(index = range(0,6)).drop(columns='Unnamed: 9', axis=1)
judge_4_clean = judge_4.drop(index = range(0,21)).drop(columns='Unnamed: 9', axis=1)
judge_5_clean = judge_5.drop(index = range(0,25)).drop(columns='Unnamed: 9', axis=1)
judge_6_clean = judge_6.drop(index = range(0,10)).drop(columns='Unnamed: 9', axis=1)
judge_7_clean = judge_7.drop(index = range(0,14)).drop(columns='Unnamed: 9', axis=1)


judge_1_clean = judge_1_clean.iloc[:11]
judge_2_clean = judge_2_clean.iloc[:10]
judge_3_clean = judge_3_clean.iloc[:10]
judge_4_clean = judge_4_clean.iloc[:10]
judge_5_clean = judge_5_clean.iloc[:10]
judge_6_clean = judge_6_clean.iloc[:10]
judge_7_clean = judge_7_clean.iloc[:10]

judge_3_clean
```

```python
all_dataframes = [judge_1_clean, judge_2_clean, judge_3_clean, judge_4_clean, judge_5_clean, judge_6_clean, judge_7_clean]

for df in all_dataframes:
    df['Innovation & Technical Merit (10 points)'] = pd.to_numeric(df['Innovation & Technical Merit (10 points)'], errors='coerce')
    df['Impact & Relevance (10 points)'] = pd.to_numeric(df['Impact & Relevance (10 points)'], errors='coerce') 
    df['User Experience & Accessibility (10 points)'] = pd.to_numeric(df['User Experience & Accessibility (10 points)'], errors='coerce') 
    df['Ethical Considerations (10 points)'] = pd.to_numeric(df['Ethical Considerations (10 points)'], errors='coerce') 
    df['Feasibility & Implementation (10 points)'] = pd.to_numeric(df['Feasibility & Implementation (10 points)'], errors='coerce') 
    df['Total Score (50 possible)'] = pd.to_numeric(df['Total Score (50 possible)'], errors='coerce') 

combined = pd.concat(all_dataframes)

combined['Comments to help explain your scores.'] = combined['Comments to help explain your scores.'].fillna('').astype(str)
combined['Recommendations for the team to help them improve upon their prototype.'] = combined['Recommendations for the team to help them improve upon their prototype.'].fillna('').astype(str)

combined_agg_comments = combined.groupby('Project Name').agg({
    'Innovation & Technical Merit (10 points)': 'mean',  
    'Impact & Relevance (10 points)': 'mean',
    'User Experience & Accessibility (10 points)': 'mean',
    'Ethical Considerations (10 points)': 'mean',
    'Feasibility & Implementation (10 points)': 'mean',
    'Total Score (50 possible)': 'mean',
    'Comments to help explain your scores.': ' '.join,
    'Recommendations for the team to help them improve upon their prototype.': ' '.join
})
combined_agg_comments

combined_agg_comments = combined_agg_comments.round(2)

combined_agg_comments = combined_agg_comments.reset_index()
combined_agg_comments.to_csv("cleaned_judge_comments_2024.csv")
```
