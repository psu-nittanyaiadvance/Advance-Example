```python
import pandas as pd

judge_comments_raw = pd.read_csv("2023-judge-comments.csv")

print(judge_comments_raw.columns)
print(judge_comments_raw.head)
```

```python
dropped_columns = ['Completion time', 'Your Name']

judge_comments = judge_comments_raw.drop(columns=dropped_columns)

average_scores = judge_comments.groupby('Proposal Being Reviewed', as_index=False)[judge_comments.select_dtypes(include='int').columns].mean()
all_comments = judge_comments.groupby('Proposal Being Reviewed', as_index=False).agg({
    'Impact - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Feasibility - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Implementation & Scaling - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Team Capabilities - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Technical Sophistication - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Use of AI & ML Technologies - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Use of Available Data - Comments': lambda x: ', '.join(x.dropna().astype(str),),
    'Interface Design Plans/Consumability - Comments': lambda x: ', '.join(x.dropna().astype(str),)
})

scores_and_comments = pd.merge(average_scores, all_comments)
print(scores_and_comments.head())

scores_and_comments.to_csv("cleaned_judge_comments.csv")
```

```python
submission = ['CareerUP', 'Intern Quest', 'Earality', 'Therapeutrack', 'TransferMaster', 'Trinity']
filtered_submission = scores_and_comments[scores_and_comments['Proposal Being Reviewed'].isin(submission)]

filtered_submission['Total'] = filtered_submission.select_dtypes(include = 'float').sum(axis=1)

print(filtered_submission.head)
```

```python
with open('2023-judge-comments.txt', 'w') as file:
    for index, row in filtered_submission.iterrows():
        for col in filtered_submission.columns:
            file.write(f"{col}: {row[col]}\n")
        file.write('\n')
```
